## Introduction: What Is Media Streaming in Practice?

> Multimedia that is constantly received by and presented to an end-user while being delivered by a provider.

- Providing **multimedia information** to the user, usually on request.
- **Transmission of compressed multimedia data** over the **Internet**.

## Two Types of Access to Streaming Media

### Live (Broadcasting)
- Video **captured** on the **server**.
- All users watch the same current video.
- Example: **live** concert streaming.

### On Request (On Demand)
- Video **stored** on the **server**.
- Each user chooses a video to watch.
- Example: **VOD** movie streaming.

## Introducing Real-Time Transport Protocol (RTP)

- Transport layer **protocol** for real-time streaming media.
- Not a standalone protocol.
- Requiring another transport protocol to combine with:
  - Theoretically: any transport protocol
  - Practically: used only with **User Datagram Protocol (UDP)**
- Inheriting **UDP** features, including support for both **Unicast** and **Multicast** transmissions.
- **UDP+RTP** is a good multimedia transport protocol.
- **RTP** header located between **UDP** header and **UDP** payload, but not another network layer.


## RTP and Quality of Service (QoS)

- No guarantee for **QoS** in **RTP**.
- QoS guarantees through cooperation with other protocols, such as:
  - **Real-Time Streaming Protocol (RTSP)**
  - **Session Initiation Protocol (SIP)**
  - **H.323**
  - **Resource Reservation Protocol (RSVP)**
- Their task is to establish a connection before data can be transmitted using RTP.


## RTP Limited “Real-Time”

- <span style="color:#A71930; font-weight:700;">Does not</span> reserve resources
- <span style="color:#A71930; font-weight:700;">Does not</span> guarantee QoS for real-time services, including:
  - Packet delivery (at all)
  - Scheduled packet delivery (even if lower layers do guarantee it)
  - Ordered packet delivery
- <span style="color:#00693C; font-weight:700;">But extends UDP</span> with information on:
  - Type of transmitted data
  - Timestamp
  - Serial number (of packet)


## RTP Sequential Number

- Sending packet order <span style="color:#00693C; font-weight:700;">possible</span> to reconstruct at receiver.
- <span style="color:#00693C; font-weight:700;">Possible</span> to correctly localize packet in stream, for example without decompressing video frames.


## Applications of RTP

- **Voice over IP (VoIP)**
- **Video on Demand (VoD)**
- Other applications of audio/video
- Distributed simulations
- Games
- Monitoring
- Other applications having access to real-time data


## RTP Control Protocol (RTCP)

- Not to be confused with <span style="color:#FF0000; font-weight:700;">Real Time Streaming Protocol (RTSP)</span>.
- <span style="color:#00B0F0; font-weight:700;">RTP</span> transport controlled by <span style="color:#FF0000; font-weight:700;">RTCP</span>.
- Main additional functionality by <span style="color:#FF0000; font-weight:700;">RTCP</span>: QoS monitoring for volumes of data transports.
- <span style="color:#00B0F0; font-weight:700;">RTP</span> and <span style="color:#FF0000; font-weight:700;">RTCP</span> working simultaneously.

## Real Time Streaming Protocol (RTSP)

- Application layer **protocol** to control **streaming media server**.
- “Network remote control” for **multimedia servers** and more.
- Negotiating:
  - Communication (**Unicast**, **Multicast** …)
  - Transport protocols (**UDP**, **RTP**, **HTTP** …)
  - Codecs…
- Syntax similar to **HTTP**, but supports states.

## Transcoder

- Various technical limitations for:
  - <span style="color:#A71930; font-weight:700;">Screen size and type</span>
  - <span style="color:#A71930; font-weight:700;">Processor</span>
  - <span style="color:#A71930; font-weight:700;">Memory for buffering</span>
  - <span style="color:#A71930; font-weight:700;">Communications</span>
  - <span style="color:#A71930; font-weight:700;">Power source</span>
- <span style="color:#A71930; font-weight:700;">No</span> universal digital video signal directly available for all kinds of devices.
- Solution: <span style="color:#00693C; font-weight:700;">transcoder</span>.
